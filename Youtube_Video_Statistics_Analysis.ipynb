```python
import datetime

original_format='%y.%d.%m'
desired_format='%d-%m-%y'
def convert_date(val,original_format,desired_format):
    datetime_obj = datetime.datetime.strptime(val, original_format)
    formatted_date = datetime_obj.strftime(desired_format)
    return formatted_date

df['trending_date']=df['trending_date'].apply(lambda x: convert_date(x, original_format, desired_format))
```

```python
import json

def map_categories(df, category_mapping_file):
    # Load the JSON file containing category mappings into a dictionary
    with open(category_mapping_file, 'r') as f:
        category_mapping = json.load(f)

    # Create a dictionary to map category IDs to category names
    id_to_category = {}
    for category_data in category_mapping['items']:
        category_id = category_data['id']
        category_name = category_data['snippet']['title']
        id_to_category[int(category_id)] = category_name

    # Map category IDs to category names in the DataFrame
    df['category_name'] = df['category_id'].map(id_to_category)

    return df

category_mapping_file = r"E:\Purvang\Data Set\CA_category_id.json" # Path to your JSON file
mapped_df = map_categories(df, category_mapping_file)
```

### Trending Tags

```python
tags_df=df['tags'].value_counts().reset_index()
tags_df.columns=['tags','counts']
tags_df
```

```python
def trending_tags(df, tags_df, col):
    for tag in df['tags'].unique():
        tag_sum = df[df['tags'] == tag][col].sum()
        tags_df.loc[tags_df['tags'] == tag, col] = tag_sum
```

```python
trending_tags(df,tags_df,'views')
```

### Exploring Distribution 

#### Ploting A Histogram 

```python
numeric_columns=['views','likes','dislikes','comment_count']
fig, axes = plt.subplots(nrows=len(numeric_columns), ncols=1, figsize=(6,10))

# Plot histograms for each numeric column
for i, col in enumerate(numeric_columns):
    df[col] = pd.to_numeric(df[col])
    print(f"Minimum value of {col}: {df[col].min()}")
    print(f"Maximum value of {col}: {df[col].max()}")
    sns.histplot(df[col], kde=False, ax=axes[i],bins=30)
    axes[i].set_xlabel(col)
    axes[i].set_title('Histogram of ' + col)
    axes[i].set_xlim(df[col].min(), df[col].max())
    axes[i].set_ylim(0, None)
# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

#### Ploting  A KDE Plot

```python
numeric_columns=['views','likes','dislikes','comment_count']
fig, axes = plt.subplots(nrows=len(numeric_columns), ncols=1, figsize=(6,10))

# Plot histograms for each numeric column
for i, col in enumerate(numeric_columns):
    df[col] = pd.to_numeric(df[col])
    print(f"Minimum value of {col}: {df[col].min()}")
    print(f"Maximum value of {col}: {df[col].max()}")
    sns.kdeplot(df[col],ax=axes[i])
    axes[i].set_xlabel(col)
    axes[i].set_title('kde of ' + col)
    axes[i].set_xlim(df[col].min(), df[col].max())
    axes[i].set_ylim(0, None)
# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

### Insights from YouTube Video Statistics Analysis:
1. **Time to Trending:**
   - It takes approximately 1 day for a video to become trending after posting.
   - Maximum views were recorded in December 2017, while minimum views were observed in February 2018.

2. **Effect of Tags on Views:**
   - Videos with no tags received the maximum views.
   - However, excluding videos with no tags, the top 3 tags contributing to views are:
     - 'Rewind|"Rewind 2017"|"youtube rewind 2017"|"#Y...' (Total views: 631,428,300)
     - 'marvel|"comics"|"comic books"|"nerdy"|"geeky"|...' (Total views: 577,553,200)
     - 'Childish Gambino|"Rap"|"This Is America"|"mcDJ...' (Total views: 411,775,100)

3. **Top Performing Channels:**
   - Channels with the maximum views: Marvel Entertainment, T-Series, Dude Perfect
   - Channels with the maximum likes: ibighit, PewDiePie, Marvel Entertainment
   - Channels with the maximum dislikes: YouTube Spotlight, Jake Paul, Logan Paul Vlogs
   - Channels with the maximum comment count: ibighit, YouTube Spotlight, Logan Paul Vlogs

4. **Category Analysis:**
   - Categories with the maximum views: Entertainment, Music, Comedy
   - Categories with the maximum likes: Music, Entertainment, Comedy
   - Categories with the maximum dislikes: Entertainment, Music, People & Blogs
   - Categories with the maximum comment count: Entertainment, Music, Comedy
